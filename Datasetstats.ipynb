```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size': 22})
```

```python
cities = ['Oslo', 'Bergen', 'Trondheim', 'Stavanger', 'Tromsø', 'Bodø']
areas = np.array([198.7, 139, 88.6, 35.8, 42.0, 13.7])
building_labels = np.array([17.47, 11.96, 6.93, 4.13, 2.45, 1.43])
road_labels = np.array([9.93, 7.12, 5.13, 2.84, 1.66, 0.96])
vegetation_labels = np.array([64.73, 56.13, 40.52, 6.32, 11.20, 4.97])
water_labels = np.array([15.71, 26.92, 12.61, 9.32, 18.73, 1.69])
```

```python
building_per_area = building_labels / areas
road_per_area = road_labels / areas
vegetation_per_area = vegetation_labels / areas
water_per_area = water_labels / areas
```

```python

n = 6
width = 0.20
ind = np.arange(n)
fig, ax = plt.subplots()
arr = building_per_area
arr1 = road_per_area 
arr2 = vegetation_per_area 
arr3 = water_per_area
ax.grid(True )

rects1 = ax.bar(ind, arr, width)
rects2 = ax.bar(ind + width, arr1, width)
rects3 = ax.bar(ind + width*2, arr2, width)
rects4 = ax.bar(ind + width*3, arr3, width)


# add some text for labels, title and axes ticks
ax.set_ylabel('Label area / total city area')
ax.set_title('Label distribution')
ax.set_xticks(ind + width*3 / 2)
ax.set_xticklabels(cities)
ax.legend((rects1[0], rects2[0], rects3[0], rects4[0]), ('Buildings', 'Roads', 'Vegetation', 'Water'))
```
